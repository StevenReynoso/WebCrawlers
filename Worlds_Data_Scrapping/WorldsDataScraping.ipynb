```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
%matplotlib inline
```

```python
from urllib import request
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = "https://lol.fandom.com/wiki/Special:RunQuery/TournamentStatistics?TS%5Btournament%5D=2023+Season+World+Championship%2FMain+Event&TS%5Bpreload%5D=TournamentByPlayer&_run=#pfForm"
html = urlopen(url)
```

```python
soup = BeautifulSoup(html, 'lxml')
```

```python
BeautifulSoup
```

```python
title = soup.title
print(title)
```

```python
# Find the table in the HTML
table = soup.find('table')
```

```python
# Extract column names (header)
header = [th.get_text(strip=True) for th in table.find_all('th')]
```

```python
# Create an empty list to store the data rows
data_rows = []
```

```python
# Extract and clean each row
for row in table.find_all('tr')[1:]:  # Skip the first row since it contains the header
    row_data = [td.get_text(strip=True) for td in row.find_all('td')]
    data_rows.append(row_data)
```

```python
text = soup.get_text()
```

```python
all_links = soup.find_all("a")
for link in all_links:
    print(link.get("href"))
```

```python
rows = soup.find_all('tr')
for row in rows:
    row_td = row.find_all('td')
print(row_td)
type(row_td)
```

```python
# Cleaning the html tags using beautiful soup
str_cells = str(row_td)
cleantext = BeautifulSoup(str_cells, "lxml").get_text()
print(cleantext)
```

```python
# Create an empty list to store the cleaned rows
list_rows = []
```

```python
# Extract and clean each row
for row in rows:
    cells = row.find_all('td')
    str_cells = [str(cell.get_text(strip=True)) for cell in cells]
    list_rows.append(str_cells)
```

```python
# Create a DataFrame from the list of cleaned rows
df1 = pd.DataFrame(list_rows)
df1.head(15)
```

```python
# Assuming you have already extracted the column labels
col_labels = soup.find_all('th')
all_header = []

# Extract and split the column labels
for th in col_labels:
    label_text = th.get_text(strip=True)
    labels = label_text.split(', ')  # Split by comma and space
    all_header.extend(labels)

# Now, all_header will contain the column labels as separate elements
print(all_header)
```

```python
# fixing headers to start where we need it to for the dataset
new_columns = ['Team', 'Player']
start_index = all_header.index('G')
column_labels = new_columns + all_header[start_index:]
print(column_labels)
```

```python
df3 = pd.DataFrame(column_labels)
df3 = df3.T
df3.head()
```

```python
frames = [df3, df1]
df4 = pd.concat(frames)
df4.head()
```

```python
df5 = df4.rename(columns=df4.iloc[0])
df5.head(15)
```

```python
# remove extra empty rows in dataset
start_index = 6
selected_rows = df5[start_index:]
selected_rows.head()
```
